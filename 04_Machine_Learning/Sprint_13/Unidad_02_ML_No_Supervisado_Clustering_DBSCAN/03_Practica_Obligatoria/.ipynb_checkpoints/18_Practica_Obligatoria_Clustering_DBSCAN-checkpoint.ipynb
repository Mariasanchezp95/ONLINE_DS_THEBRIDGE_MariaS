### Ejercicio 3 — Tratamiento de features (escalado)

Aunque las variables del dataset ya presentan medias cercanas a 0 y desviaciones típicas próximas a 1, es recomendable aplicar un nuevo escalado antes de utilizar algoritmos basados en distancias como DBSCAN o el clustering jerárquico. Estos métodos son muy sensibles a las diferencias de escala entre variables, por lo que garantizar que todas contribuyen por igual es fundamental.

Para ello utilizamos `StandardScaler`, que transforma cada variable restando su media y dividiéndola por su desviación típica. El resultado es un dataset completamente estandarizado y listo para el modelado.

Tras aplicar `StandardScaler`, comprobamos que todas las variables quedan centradas en 0 y con desviación típica 1. Aunque los datos originales ya parecían normalizados, este paso garantiza que todas las features contribuyen de forma homogénea a las distancias utilizadas por DBSCAN y el clustering jerárquico. Esto es especialmente importante porque ambos algoritmos son muy sensibles a diferencias de escala entre variables.

Además, el escalado permite que los hiperparámetros de DBSCAN (especialmente `eps`) sean más interpretables y comparables entre variables, evitando que una feature domine la distancia euclídea.

```python
# 4. DBSCAN: búsqueda de eps y entrenamiento del modelo

from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Número de features del dataset escalado
n_features = df_scaled.shape[1]

# min_samples recomendado: 2 × número de features
min_samples_value = 2 * n_features

# Lista para guardar el porcentaje de outliers para cada eps
outlier_percentages = []

# Probamos valores de eps entre 0.1 y 3 (50 valores)
eps_values = np.linspace(0.1, 3, 50)

for eps in eps_values:
    # Creamos el modelo DBSCAN con el eps actual
    dbscan = DBSCAN(eps=eps, min_samples=min_samples_value)
    
    # Ajustamos el modelo a los datos escalados
    dbscan.fit(df_scaled)
    
    # Calculamos el porcentaje de outliers (etiqueta = -1)
    outliers = np.sum(dbscan.labels_ == -1)
    percentage = 100 * outliers / len(dbscan.labels_)
    
    outlier_percentages.append(percentage)

# --------------------------------------------------------------
# Gráfico del porcentaje de outliers según eps
# --------------------------------------------------------------

plt.figure(figsize=(10, 5))
sns.lineplot(x=eps_values, y=outlier_percentages)
plt.xlabel("Valor de eps")
plt.ylabel("Porcentaje de outliers (%)")
plt.title("Selección de eps para DBSCAN")
plt.grid(True)
plt.show()

# --------------------------------------------------------------
# Selección del eps óptimo
# --------------------------------------------------------------
# Buscamos el primer eps cuyo porcentaje de outliers sea < 10%
eps_optimo = None
for eps, perc in zip(eps_values, outlier_percentages):
    if perc < 10:
        eps_optimo = eps
        break

print("Eps recomendado:", eps_optimo)

# --------------------------------------------------------------
# Entrenamiento final del modelo DBSCAN con eps óptimo
# --------------------------------------------------------------

dbscan_final = DBSCAN(eps=eps_optimo, min_samples=min_samples_value)
dbscan_final.fit(df_scaled)

# Etiquetas obtenidas
labels = dbscan_final.labels_

# Conteo de clusters
unique, counts = np.unique(labels, return_counts=True)
cluster_summary = dict(zip(unique, counts))

print("\nResumen de clusters encontrados:")
print(cluster_summary)
```

### Ejercicio 4 — DBSCAN: selección de hiperparámetros y entrenamiento

Para aplicar DBSCAN necesitamos seleccionar dos hiperparámetros clave:
- **eps**: radio del vecindario
- **min_samples**: número mínimo de puntos para considerar un punto como “core”

Siguiendo la recomendación habitual, fijamos:[text{min\_samples} = 2 \times \text{número de features} = 10]

A continuación probamos distintos valores de `eps` entre 0.1 y 3, midiendo el porcentaje de outliers (puntos etiquetados como -1). El objetivo es encontrar un valor de `eps` que deje menos del 10% de outliers, siguiendo el criterio utilizado en la Unidad 13.

Tras analizar la curva, seleccionamos el primer valor de `eps` que cumple esta condición y entrenamos el modelo DBSCAN definitivo. Finalmente mostramos cuántos clusters ha encontrado el algoritmo y cuántos puntos pertenecen a cada uno.

Este proceso es especulativo porque no existe una métrica universal para evaluar la “calidad” de un clustering sin etiquetas reales. La elección de eps se basa en criterios heurísticos y en la interpretación visual del comportamiento del algoritmo.

### Resultados de DBSCAN

Tras realizar la búsqueda de hiperparámetros, el valor recomendado de `eps` fue aproximadamente **1.28**, siendo el primer valor que dejaba menos del 10% de outliers. Con este valor y `min_samples = 10`, entrenamos el modelo DBSCAN final.

El algoritmo encontró:
- **1 cluster principal** con 140 mezclas
- **13 outliers** etiquetados como `-1`

Esto indica que las mezclas químicas forman un grupo muy compacto en el espacio de características, sin subgrupos densos claramente diferenciados. DBSCAN interpreta que la mayoría de las bebidas pertenecen a una misma región de alta densidad, mientras que unas pocas combinaciones químicas aparecen como casos atípicos.

Este resultado es razonable, ya que las variables están estandarizadas y no existen separaciones fuertes entre mezclas. Los outliers representan formulaciones químicas menos frecuentes o más extremas. 

Aún así, la interpretación es especulativa porque no existe una variable objetivo que permita validar si la segmentación tiene sentido desde el punto de vista del sabor o de la eficacia antifatiga. El clustering se basa únicamente en la estructura matemática de los datos, no en preferencias reales de usuarios.

```python
# Ejercicio 5 — Clustering Jerárquico

# En este ejercicio aplicamos clustering jerárquico para identificar
# posibles grupos de mezclas químicas. Este método permite visualizar
# la estructura de los datos mediante un dendrograma.

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# --------------------------------------------------------------
# 1. Construcción del dendrograma
# --------------------------------------------------------------
# Usamos el método 'average' porque es más estable que 'single' o 'complete'
# y no asume clusters esféricos como 'ward'.

linkage_method = linkage(df_scaled, method='average')

plt.figure(figsize=(14, 6))
dendrogram(linkage_method)
plt.title("Dendrograma — Clustering Jerárquico (linkage = average)")
plt.xlabel("Muestras")
plt.ylabel("Distancia")
plt.show()

# --------------------------------------------------------------
# 2. Selección del número de clusters
# --------------------------------------------------------------
# Observando el dendrograma, elegimos un número razonable de clusters.
# La práctica recomienda entre 3 y 5, así que seleccionamos 4 como valor inicial.

n_clusters = 4

# --------------------------------------------------------------
# 3. Entrenamiento del modelo AgglomerativeClustering
# --------------------------------------------------------------

model_hier = AgglomerativeClustering(
    n_clusters=n_clusters,
    metric='euclidean',
    linkage='average'
)

labels_hier = model_hier.fit_predict(df_scaled)

# --------------------------------------------------------------
# 4. Análisis de los clusters obtenidos
# --------------------------------------------------------------

# Añadimos las etiquetas al DataFrame original
df_hier = df.copy()
df_hier["Cluster_Hier"] = labels_hier

# Conteo de elementos por cluster
cluster_counts = df_hier["Cluster_Hier"].value_counts().sort_index()
print("Número de mezclas por cluster:")
print(cluster_counts)

# --------------------------------------------------------------
# 5. Perfil medio de cada cluster
# --------------------------------------------------------------

cluster_means = df_hier.groupby("Cluster_Hier").mean()
print("\nValores medios por cluster:")
print(cluster_means)
```

### Ejercicio 5 — Clustering Jerárquico

Aplicamos clustering jerárquico para identificar posibles grupos de mezclas químicas. Este método es especialmente útil porque permite visualizar la estructura de los datos mediante un dendrograma, lo que facilita la elección del número de clusters.

#### 1. Dendrograma
Construimos el dendrograma utilizando el método de enlace *average*, que calcula la distancia entre clusters como la media de las distancias entre sus elementos. Este método es más estable que *single* o *complete* y no asume clusters esféricos como *ward*.

El dendrograma permite observar a qué alturas se fusionan los clusters. Siguiendo la recomendación del enunciado (entre 3 y 5 clusters), seleccionamos **4 clusters** como punto de corte razonable.

#### 2. Entrenamiento del modelo
Entrenamos un modelo `AgglomerativeClustering` con:
- `n_clusters = 4`
- `metric = euclidean`
- `linkage = average`

#### 3. Resultados
Tras asignar las etiquetas, analizamos cuántas mezclas pertenecen a cada cluster y calculamos los valores medios de cada componente químico dentro de cada grupo. Esto permite interpretar qué caracteriza a cada segmento de bebidas.

Este análisis es especulativo porque no existe una variable objetivo que indique si un cluster corresponde a una bebida más eficaz, más agradable o más energizante. La segmentación se basa únicamente en la estructura matemática de los datos.

### **Comparación entre DBSCAN y Clustering Jerárquico**

En esta práctica hemos aplicado dos algoritmos de clustering no supervisado: **DBSCAN** y **Clustering Jerárquico**. Ambos buscan segmentar las mezclas químicas sin disponer de una variable objetivo, pero lo hacen desde enfoques muy distintos.

#### 1. Resultados de DBSCAN
DBSCAN encontró:
- **1 único cluster real** (140 mezclas)
- **13 outliers**

Esto indica que las mezclas forman un grupo muy compacto en términos de densidad. DBSCAN no detecta subgrupos internos porque no existen regiones de alta densidad separadas entre sí. Los 13 outliers representan formulaciones químicas poco frecuentes o extremas.

#### 2. Resultados del Clustering Jerárquico
El clustering jerárquico con 4 clusters reveló:
- **Dos grupos grandes** (88 y 63 mezclas)
- **Dos mezclas extremas** que forman clusters individuales

Este método sí detecta variaciones internas en la estructura de los datos, separando mezclas “suaves” y “fuertes”, además de dos formulaciones muy atípicas.

#### 3. ¿Por qué los resultados son diferentes?
- **DBSCAN** se basa en densidades. Si no hay zonas densas claramente separadas, agrupa todo en un único cluster.
- **El clustering jerárquico** se basa en distancias y siempre fuerza una estructura jerárquica, lo que permite identificar subgrupos incluso cuando no hay densidades diferenciadas.

#### 4. ¿Cuál es más útil en este caso?
Depende del objetivo:

- Si los químicos quieren **detectar mezclas atípicas**, DBSCAN es útil porque identifica outliers de forma natural.
- Si quieren **segmentar el catálogo en grupos interpretables**, el clustering jerárquico es más informativo porque revela niveles de intensidad química.

#### 5. ¿Por qué la comparación es especulativa?
Porque no existe una variable objetivo que indique qué segmentación es “mejor”.  
Ambos algoritmos agrupan basándose únicamente en la estructura matemática de los datos, no en preferencias reales de usuarios ni en eficacia antifatiga. Por tanto, cualquier interpretación depende del contexto químico y no puede validarse empíricamente.

### Conclusión
- **DBSCAN** → un único cluster + outliers  
- **Jerárquico** → dos clusters principales + dos mezclas extremas  
- Ambos métodos aportan información complementaria y ayudan a los químicos a entender la variabilidad de las formulaciones.

### Parte II — Clasificación de nuevos experimentos

Los químicos proporcionan un nuevo conjunto de mezclas experimentales y solicitan que las clasifiquemos según los segmentos obtenidos en la Parte I. Como DBSCAN no permite predecir nuevos puntos, utilizamos el modelo de clustering jerárquico.

El procedimiento es el siguiente:

1. Cargamos los nuevos experimentos y aplicamos el mismo preprocesado que al dataset original.
2. Escalamos los datos utilizando el mismo `StandardScaler` ajustado previamente.
3. Calculamos la distancia de cada nuevo experimento a los centroides de los clusters jerárquicos.
4. Asignamos cada experimento al cluster cuyo centroide esté más cercano.

Este método es estándar cuando se quiere extender un clustering jerárquico a nuevos datos. El resultado final indica a qué segmento pertenece cada nueva mezcla, permitiendo a los químicos interpretar rápidamente si se trata de una formulación suave, fuerte o extrema.

Tras escalar los nuevos experimentos y asignarlos al cluster más cercano mediante la distancia a los centroides del clustering jerárquico, obtenemos la siguiente clasificación:

- **Experimentos 0 y 1 → Cluster 1**
- **Experimentos 2, 3, 4 y 5 → Cluster 0**

Recordando la interpretación de la Parte I:

#### **Cluster 1 — Mezclas fuertes o de alta intensidad**
Los experimentos 0 y 1 presentan:
- Valores altos en cafeína, vitaminas y azúcares
- Mayor concentración general de componentes
- Perfiles similares a las formulaciones más potentes del dataset original

Estas mezclas podrían corresponder a prototipos con efectos energéticos más intensos.

#### **Cluster 0 — Mezclas suaves o de baja intensidad**
Los experimentos 2, 3, 4 y 5 muestran:
- Valores más bajos en la mayoría de componentes
- Perfiles más moderados y menos concentrados
- Similitud con las formulaciones suaves del dataset original

Estas mezclas serían candidatas a bebidas menos intensas, posiblemente más equilibradas o de sabor más suave.

### Conclusión
Los nuevos experimentos se distribuyen en dos segmentos claros:
- **2 mezclas fuertes (Cluster 1)**
- **4 mezclas suaves (Cluster 0)**

Esta clasificación permite a los químicos priorizar qué prototipos explorar según el tipo de bebida que quieran desarrollar: más potente o más ligera.
