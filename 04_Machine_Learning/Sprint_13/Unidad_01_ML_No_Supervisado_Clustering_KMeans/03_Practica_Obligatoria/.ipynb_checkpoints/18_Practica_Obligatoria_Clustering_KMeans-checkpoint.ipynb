### #1.3

Vamos a dividir en train y test, pero OJO RECUERDA QUE EN LOS PROBLEMAS NO SUPERVISADOS NO HAY SPLIT (porque no hay target), aquí lo hacemos para poder comparar posteriormente el clustering con la clasificación (ya que es uan práctica formativa).

Por tanto, divídelo en un conjunto de entrenamiento, un conjunto de validación y un conjunto de pruebas (80-10-10). Dado que el conjunto de datos es bastante pequeño, emplea un muestreo estratificado para asegurarse de que haya el mismo número de imágenes por persona en cada conjunto (estratificando por la columna que contenga el target)

NOTA: No hemos hecho hasta ahora la separación en tres sets, investiga por tu cuenta o bien haz primero un split 90-10 y luego otro split 89-11 sobre el de 80 (para que de los números aproximados), por ejemplo.

````python
from sklearn.model_selection import train_test_split

# --------------------------------------------------------------
# Primer split: 90% train_val, 10% test
# --------------------------------------------------------------
train_val, test = train_test_split(
    df_shuffled,
    test_size=0.10,
    stratify=df_shuffled["target"],
    random_state=42
)

# --------------------------------------------------------------
# Segundo split: del 90%, separar 80% train y 10% valid
# (80/90 ≈ 0.888..., por eso usamos test_size=0.111)
# --------------------------------------------------------------
train, valid = train_test_split(
    train_val,
    test_size=0.111,   # para que valid ≈ 10% del total
    stratify=train_val["target"],
    random_state=42
)

# Mostrar tamaños
len(train), len(valid), len(test)
````

### Ejercicio 1.3 — División en train, valid y test (80-10-10)

En los problemas no supervisados normalmente no se realiza un split, una división en conjuntos de entrenamiento y prueba, ya que no existe un "target" que predecir. Sin embargo, en esta práctica sí hacemos un split porque más adelante compararemos el resultado del clustering con un modelo supervisado. Por este motivo, necesitamos separar los datos en tres subconjuntos: entrenamiento, validación y prueba.

El dataset Olivetti contiene 400 imágenes correspondientes a 40 personas distintas, con 10 imágenes por persona. Para mantener la misma proporción de imágenes por persona en cada conjunto, es necesario realizar un muestreo **estratificado** usando la columna `target`.

Como `train_test_split` no permite crear tres conjuntos directamente, realizamos dos divisiones:
1. **Un primer split 90%-10%** para separar el conjunto de test: Separamos un 10% del dataset para el conjunto de prueba (`test`). El 90% restante se mantiene como `train_val`.
2. **Un segundo split sobre el 90% restante** para obtener train (80%) y validación (10%). Dividimos `train_val` en:
   - un 80% del total para entrenamiento (`train`)
   - un 10% del total para validación (`valid`)
     Para conseguir esta proporción, dividimos `train_val` usando `test_size=0.111`.
     
De esta forma obtenemos tres conjuntos estratificados con proporciones aproximadas 80-10-10. Gracias a la estratificación, cada conjunto mantiene el mismo número de imágenes por persona, evitando sesgos en la distribución.

El resultado final es:
- **320 imágenes** en el conjunto de entrenamiento (80%)
- **40 imágenes** en el conjunto de validación (10%)
- **40 imágenes** en el conjunto de prueba (10%)

Esta división equilibrada nos permitirá evaluar posteriormente el rendimiento del clustering comparándolo con un modelo supervisado.

Utiliza la siguiente función para visualizar alguna de las caras (observa que tienes que dar la X y la y, usa iloc en ambos datasets)

````python
def plot_faces(faces, labels, n_cols=5):
    faces = faces.reshape(-1, 64, 64)
    n_rows = (len(faces) - 1) // n_cols + 1
    plt.figure(figsize=(n_cols, n_rows * 1.1))
    for index, (face, label) in enumerate(zip(faces, labels)):
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(face, cmap="gray")
        plt.axis("off")
        plt.title(label)
    plt.show()
````

````python
# Visualizar algunas caras del train (ej: 10 primeras)
plot_faces(X_train[:10], y_train[:10])
````

PCA permite proyectar los datos en un espacio de menor dimensión conservando la mayor parte de la información original.

En este ejercicio utilizamos:

```python
pca = PCA(0.99)

El modelo PCA se ajusta únicamente con el conjunto de entrenamiento (X_train) para evitar fugas de información, y posteriormente se utiliza para transformar los conjuntos de validación y prueba.

### 1.6

Aquí viene la parte del león. A continuación, agrupa las imágenes utilizando K-Means sobre el dataset de train reducido en el ejercicio anterior. Emplea el método del máximo de score de silueta para obtener el mejor k, probando con K de 5 en 5 hasta 150. ¿Cuál es el k que proporciona el mejor score de silueta? NOTA: Emplea todas las features (no hace falta seleccionar, y ya están escaladas entre 0 y 1)

````python
# Importamos KMeans y la métrica de silueta
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Definimos los valores de k que queremos probar (de 5 en 5 hasta 150)
ks = range(5, 151, 5)

# Lista para guardar los valores del silhouette score
sil_scores = []

# Bucle para probar cada valor de k
for k in ks:
    # Creamos el modelo K-Means con k clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Ajustamos el modelo y obtenemos las etiquetas asignadas
    labels = kmeans.fit_predict(X_train_pca)
    
    # Calculamos el silhouette score para este k
    score = silhouette_score(X_train_pca, labels)
    
    # Guardamos el resultado
    sil_scores.append(score)

# Obtenemos el mejor k (el que maximiza el silhouette score)
best_k = ks[np.argmax(sil_scores)]

# Mostramos el mejor k y su score
best_k, max(sil_scores)
````

### Ejercicio 1.6 — Selección del mejor número de clusters con el método del máximo Silhouette Score

En este ejercicio aplicamos el algoritmo K-Means sobre los datos reducidos mediante PCA (`X_train_pca`). El objetivo es determinar cuál es el número óptimo de clusters `k` utilizando el **Silhouette Score**, una métrica que evalúa la calidad de la separación entre clusters.

El Silhouette Score toma valores entre -1 y 1:
- valores cercanos a **1** indican clusters bien definidos,
- valores cercanos a **0** indican solapamiento entre clusters,
- valores negativos indican asignaciones incorrectas.

Para encontrar el mejor `k`, probamos valores desde **5 hasta 150**, avanzando de 5 en 5. Para cada valor de `k`:
1. Entrenamos un modelo K-Means sobre `X_train_pca`.
2. Obtenemos las etiquetas asignadas a cada imagen.
3. Calculamos el Silhouette Score correspondiente.
4. Guardamos el resultado para compararlo posteriormente.

Finalmente, seleccionamos el `k` que produce el **mayor Silhouette Score**, lo que indica la mejor estructura de clustering según esta métrica. En nuestro caso, el valor de k que maximiza el Silhouette Score es **k = 110**, con un score aproximado de **0.1989**.


Este proceso nos permite elegir un número de clusters adecuado sin necesidad de inspección visual, aprovechando la información geométrica de los datos reducidos con PCA.

### #1.7

Repite el ejercio anterior empleando ahora el método del codo de Inercia. ¿Sale algo más concluyente o que refuerce el anterior resultado?

````python
# Importamos KMeans
from sklearn.cluster import KMeans

# Lista para guardar las inercias
inertias = []

# Probamos los mismos valores de k que en el ejercicio anterior
ks = range(5, 151, 5)

for k in ks:
    # Creamos el modelo K-Means con k clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    
    # Ajustamos el modelo sobre los datos reducidos con PCA
    kmeans.fit(X_train_pca)
    
    # Guardamos la inercia (suma de distancias dentro de cada cluster)
    inertias.append(kmeans.inertia_)

# Mostramos los valores de inercia para inspección
inertias
````

````python
# Visualizar el codo
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(ks, inertias, marker='o')
plt.xlabel("Número de clusters (k)")
plt.ylabel("Inercia")
plt.title("Método del codo")
plt.grid(True)
plt.show()
````

### Ejercicio 1.7 — Método del codo de Inercia

En este ejercicio repetimos el análisis del número óptimo de clusters utilizando el método del codo basado en la inercia. La inercia mide la suma de las distancias de cada punto a su centroide, por lo que valores más bajos indican clusters más compactos.

Probamos los mismos valores de `k` que en el ejercicio anterior (de 5 en 5 hasta 150) y calculamos la inercia para cada uno de ellos. Al representar la curva de inercia frente a `k`, buscamos el punto donde la disminución deja de ser pronunciada: ese punto se conoce como “el codo”.

La curva obtenida muestra un descenso rápido en los primeros valores de `k`, seguido de una disminución cada vez más suave. A partir de aproximadamente **k ≈ 100–120**, la mejora en la inercia se vuelve marginal y la curva comienza a aplanarse. Este comportamiento indica que añadir más clusters a partir de ese rango no aporta una mejora significativa en la compactación de los grupos.

Este resultado coincide con el obtenido mediante el Silhouette Score en el ejercicio anterior, donde el mejor valor fue **k = 110** con un score aproximado de **0.1989**. Ambos métodos apuntan a un número de clusters similar, lo que refuerza la conclusión de que un valor alrededor de **110 clusters** es razonable para este dataset tras la reducción de dimensionalidad con PCA.
