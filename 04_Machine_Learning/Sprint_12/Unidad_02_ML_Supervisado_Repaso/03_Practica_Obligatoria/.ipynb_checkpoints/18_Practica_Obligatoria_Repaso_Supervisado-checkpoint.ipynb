### #1 Explicación del dataset y carga de datos

Vamos a trabajar con el dataset "Wine Quality", que es un dataset en el que se recogen diversas características fisicoquímicas de las variades tinta y blanca del "Vinho verde" portugués. Las variables del dataset se describen a continuación


1. **Acidez fija**: La mayoría de los ácidos involucrados con el vino son fijos o no volátiles (no se evaporan fácilmente).
2. **Acidez volátil**: La cantidad de ácido acético en el vino, que en niveles demasiado altos puede llevar a un sabor desagradable a vinagre.
3. **Ácido cítrico**: Encontrado en pequeñas cantidades, el ácido cítrico puede añadir 'frescura' y sabor a los vinos.
4. **Azúcar residual**: Es la cantidad de azúcar restante después de que se detiene la fermentación, es raro encontrar vinos con menos de 1 gramo/litro y vinos con más de 45 gramos/litro se consideran dulces.
5. **Cloruros**: La cantidad de sal en el vino.
6. **Dióxido de azufre libre**: La porción libre de SO2 existe en equilibrio entre el SO2 molecular (como un gas disuelto) y el ion bisulfito; previene el crecimiento microbiano y la oxidación del vino.
7. **Dióxido de azufre total**: Es la cantidad de formas libres y ligadas de S02; en bajas concentraciones, el SO2 es en su mayoría indetectable en el vino, pero en concentraciones libres de más de 50 ppm, el SO2 se hace evidente en la nariz y el sabor del vino.
8. **Densidad**: La densidad del vino es cercana a la del agua dependiendo del porcentaje de alcohol y contenido de azúcar.
9. **pH**: Describe qué tan ácido o básico es un vino en una escala de 0 (muy ácido) a 14 (muy básico); la mayoría de los vinos están entre 3-4 en la escala de pH.
10. **Sulfatos**: Un aditivo del vino que puede contribuir a los niveles de dióxido de azufre (SO2), que actúa como un antimicrobiano y antioxidante.
11. **Alcohol**: El porcentaje de contenido de alcohol del vino.
12. **Calidad**: Una variable de salida (basada en datos sensoriales, puntuación entre 0 y 10).
13. **Class**: Si es blanco o tinto.



Carga el dataset "wines_dataset.csv", muestra sus primeras filas, su información general y define dos variables target una para clasificación que permita predecir la calidad del vino según el resto de propiedas y otra variable target para regresión que permita predecir el grado alcohólico del vino a partir del resto de propiedades.

### #2 Modelado para clasificación

Crea un modelo que permita clasificar lo mejor posible los vinos en función de su calidad, o dicho de otro modo que prediga de la mejor manera posible la calidad de un vino a partir de sus propiedades químicas. ***Considéralo un problema de clasificación***, donde los valores de calidad son las clases a predecir. Puedes utilizar los modelos que creas conveniente, al menos tres, y utiliza un KNN como modelo baseline, prueba al menos con 2 valores de K. Intenta mejorar el "recall medio" con las técnicas que creas conveniente. Compara los modelos utilizando validación cruzada con o sin optimización previa de hiperparámetros (aunque en la sesión en vivo por limitación de tiempos sólo lo haremos con validación cruzada). El modelo elegido si tiene que tener un ejemplo de optimización de parámetros.

Evalúa el modelo, analiza los errores y propón algún mecanismo de mejora aunque sea sólo un esbozo.

```python
df = df.drop(columns=["class"])
```

```python
# 2.1 Split en train y test para el problema de clasificación

# Definimos X e y para clasificación
X_clf = df.drop(columns=["quality"])   # todas las columnas excepto el target
y_clf = df["quality"]                  # target de clasificación

# Hacemos el split (80% train, 20% test)
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y_clf, test_size=0.2, random_state=42, stratify=y_clf
)

# Comprobamos tamaños
X_train_clf.shape, X_test_clf.shape
```

En esta celda se realiza el split del dataset para el problema de clasificación. 
Primero se separan las variables predictoras (X) del target de clasificación (y = quality). Después, se divide el dataset en train (80%) y test (20%) utilizando `train_test_split`, manteniendo la proporción de clases mediante `stratify=y_clf`. 

El resultado final es un conjunto de entrenamiento con 5.197 muestras y uno de test con 1.300 muestras, ambos con 12 variables predictoras.

```python
# 2.2 Baseline con KNN para clasificación

# Escalamos las variables numéricas porque KNN depende de distancias
scaler = StandardScaler()
X_train_clf_scaled = scaler.fit_transform(X_train_clf)
X_test_clf_scaled = scaler.transform(X_test_clf)

# Probamos dos valores de K como baseline
ks = [3, 7]

resultados_knn = {}

for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    
    # Validación cruzada usando balanced_accuracy (recall medio)
    scores = cross_val_score(
        knn,
        X_train_clf_scaled,
        y_train_clf,
        cv=5,
        scoring="balanced_accuracy"
    )
    
    resultados_knn[k] = scores.mean()

resultados_knn
```

En esta celda se ha entrenado el modelo baseline KNN probando dos valores de K (3 y 7). Antes del entrenamiento se han escalado las variables numéricas, ya que KNN se basa en distancias. 

La evaluación se ha realizado mediante validación cruzada con la métrica balanced_accuracy, que corresponde al recall medio solicitado por el negocio.

Los resultados obtenidos han sido:
- K=3 → balanced_accuracy ≈ 0.298
- K=7 → balanced_accuracy ≈ 0.274
Estos valores son bajos debido al fuerte desbalanceo del target (algunas clases tienen muy pocos ejemplos). 

El warning mostrado indica que la clase menos frecuente solo tiene 4 muestras, lo cual es insuficiente para realizar una validación cruzada de 5 folds, pero no afecta al funcionamiento del modelo.
Este baseline servirá como referencia para comparar con modelos más avanzados.

```python
# 2.3 Modelos adicionales para clasificación
# En esta celda entrenamos dos modelos: Logistic Regression y Random Forest.
# Ambos se evaluarán usando validación cruzada con balanced_accuracy, que corresponde al recall medio solicitado por el negocio.

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

# Diccionario para guardar los resultados de cada modelo
resultados_modelos = {}

# -----------------------------
# Modelo 1: Logistic Regression
# -----------------------------
# Este modelo es sensible a la escala de las variables,
# por lo que usamos los datos escalados.
log_reg = LogisticRegression(max_iter=500)

scores_log = cross_val_score(
    log_reg,
    X_train_clf_scaled,   # datos escalados
    y_train_clf,
    cv=5,
    scoring="balanced_accuracy"
)

# Guardamos el resultado medio
resultados_modelos["Logistic Regression"] = scores_log.mean()

# -----------------------------
# Modelo 2: Random Forest
# -----------------------------
# Este modelo NO necesita escalado, así que usamos los datos originales.
rf = RandomForestClassifier(random_state=42)

scores_rf = cross_val_score(
    rf,
    X_train_clf,          # datos sin escalar
    y_train_clf,
    cv=5,
    scoring="balanced_accuracy"
)

# Guardamos el resultado medio
resultados_modelos["Random Forest"] = scores_rf.mean()

# Mostramos los resultados
resultados_modelos
```

En esta celda se entrenan dos modelos adicionales para el problema de clasificación: Logistic Regression y Random Forest. La regresión logística se entrena sobre los datos escalados, ya que es un modelo sensible a la escala de las variables. Por otro lado, Random Forest no requiere escalado, por lo que se entrena directamente sobre las variables originales.

Ambos modelos se evalúan mediante validación cruzada de 5 folds utilizando la métrica balanced_accuracy, que corresponde al recall medio solicitado por el negocio. 

Los resultados obtenidos permiten comparar estos modelos con el baseline KNN y determinar cuál ofrece un mejor rendimiento inicial. Los resultados obtenidos mediante validación cruzada con balanced_accuracy son:
- Logistic Regression → ≈ 0.239
- Random Forest → ≈ 0.363

La regresión logística obtiene un rendimiento bajo debido al fuerte desbalanceo del target y a que el modelo es lineal, por lo que no captura bien las relaciones no lineales del dataset.

Por el contrario, Random Forest obtiene un rendimiento superior (arpox 0.36), lo que indica que es capaz de manejar mejor la complejidad del problema y las clases minoritarias. Este modelo se perfila como un buen candidato para la fase de optimización.



```python
# 2.4 Comparación de modelos de clasificación
# En esta celda reunimos los resultados obtenidos por todos los modelos:
# - KNN (K=3 y K=7)
# - Logistic Regression
# - Random Forest
# El objetivo es comparar sus balanced_accuracy para identificar el mejor candidato.

import pandas as pd

# Creamos un diccionario con todos los resultados
comparacion_modelos = {
    "KNN (k=3)": resultados_knn[3],
    "KNN (k=7)": resultados_knn[7],
    "Logistic Regression": resultados_modelos["Logistic Regression"],
    "Random Forest": resultados_modelos["Random Forest"]
}

# Convertimos a DataFrame para visualizarlo mejor
df_comparacion = pd.DataFrame.from_dict(
    comparacion_modelos, orient="index", columns=["balanced_accuracy"]
)

# Ordenamos de mayor a menor rendimiento
df_comparacion = df_comparacion.sort_values(by="balanced_accuracy", ascending=False)

# Mostramos la tabla comparativa
df_comparacion
```

En esta celda se comparan los resultados obtenidos por todos los modelos entrenados (KNN con k=3 y k=7, Logistic Regression y Random Forest). Para ello se construye una tabla que muestra la métrica balanced_accuracy de cada modelo y se ordena de mayor a menor rendimiento. Esta comparación permite identificar qué modelo funciona mejor con el dataset desbalanceado y cuál debe pasar a la fase de optimización.

Los resultados muestran que Random Forest es el modelo con mejor balanced_accuracy (aprox 0.363), superando claramente a KNN (aprox 0.298 y 0.274) y a Logistic Regression (aprox 0.239). Esto indica que Random Forest es el modelo que mejor maneja el desbalanceo del target y la complejidad del problema. Por tanto, se selecciona como el mejor candidato para la fase de optimización mediante GridSearchCV.

```python
# 2.5 Optimización del mejor modelo: Random Forest
# En esta celda realizamos una búsqueda de hiperparámetros usando GridSearchCV
# para mejorar el rendimiento del Random Forest.
# La métrica utilizada sigue siendo balanced_accuracy.

from sklearn.model_selection import GridSearchCV

# Definimos la rejilla de hiperparámetros a probar
param_grid = {
    "n_estimators": [100, 200],      # número de árboles
    "max_depth": [None, 5, 10],      # profundidad máxima
    "min_samples_split": [2, 5],     # mínimo de muestras para dividir un nodo
    "min_samples_leaf": [1, 2]       # mínimo de muestras por hoja
}

# Creamos el modelo base
rf = RandomForestClassifier(random_state=42)

# Configuramos GridSearchCV
grid_rf = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring="balanced_accuracy",
    n_jobs=-1
)

# Entrenamos la búsqueda
grid_rf.fit(X_train_clf, y_train_clf)

# Mostramos los mejores hiperparámetros y el score asociado
grid_rf.best_params_, grid_rf.best_score_
```

En esta celda se ha optimizado el modelo Random Forest mediante GridSearchCV. Se ha definido una rejilla de hiperparámetros que incluye el número de árboles, la profundidad máxima y los criterios mínimos para dividir nodos y hojas. 

La búsqueda se ha realizado con validación cruzada de 5 folds utilizando la métrica balanced_accuracy, manteniendo la coherencia con el criterio del negocio.

Los mejores hiperparámetros encontrados han sido:
- n_estimators = 100
- max_depth = None
- min_samples_split = 2
- min_samples_leaf = 1

El mejor balanced_accuracy obtenido durante la validación cruzada ha sido ≈ 0.363, muy similar al valor del Random Forest sin optimizar. Esto indica que el modelo ya estaba cerca de su rendimiento óptimo con los parámetros por defecto.


```python
# 2.6 Evaluación final del modelo optimizado en el conjunto de test
# En esta celda entrenamos el Random Forest con los mejores hiperparámetros
# encontrados en GridSearchCV y evaluamos su rendimiento en el conjunto de test.

from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Entrenamos el modelo final con los mejores hiperparámetros
mejor_rf = RandomForestClassifier(
    n_estimators=grid_rf.best_params_["n_estimators"],
    max_depth=grid_rf.best_params_["max_depth"],
    min_samples_split=grid_rf.best_params_["min_samples_split"],
    min_samples_leaf=grid_rf.best_params_["min_samples_leaf"],
    random_state=42
)

mejor_rf.fit(X_train_clf, y_train_clf)

# Predicciones sobre el conjunto de test
y_pred = mejor_rf.predict(X_test_clf)

# -----------------------------
# Matriz de confusión
# -----------------------------
cm = confusion_matrix(y_test_clf, y_pred)

plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Matriz de confusión - Random Forest optimizado")
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.show()

# -----------------------------
# Classification report
# -----------------------------
print("Classification Report:\n")
print(classification_report(y_test_clf, y_pred))
```

En este último punto se ha evaluado el modelo Random Forest optimizado utilizando el conjunto de test. Para ello, se entrenó el modelo final con los mejores hiperparámetros obtenidos en GridSearchCV y se generaron las predicciones sobre los datos de test.

La matriz de confusión muestra que el modelo predice correctamente las clases mayoritarias (especialmente las calidades 5 y 6), mientras que las clases con muy pocos ejemplos presentan un rendimiento bajo, algo esperable debido al fuerte desbalanceo del dataset. El classification report confirma este comportamiento: las clases minoritarias tienen valores reducidos de recall y f1-score, mientras que las clases más frecuentes alcanzan métricas más altas.

A pesar de estas limitaciones, el modelo logra una accuracy global cercana al 68% y un rendimiento razonable en balanced_accuracy dentro de las restricciones del problema. En conjunto, el Random Forest optimizado ofrece el mejor equilibrio entre precisión y capacidad para manejar el desbalanceo, consolidándose como el modelo más adecuado para este caso.

### #3 Modelado para regresión

Reutiliza todo lo que puedas la preparación del modelado de la parte anterior y construye ahora el mejor modelo que puedas para predecir el grado alcohólico de un vino dadas sus características fisico químicas, su clase y la puntuación de calidad que le ha sido otorgada. Compara al menos tres tipos de modelos, seleccionando una métrica adecuada de comparación y escoge el más apropiado empleando validación cruzada antes o después de optimizar hiperparámetros (en la sesión en vivo lo haremos antes por cuestiones de tiempo). Evalúa el modelo escogido y con los hiperparámetros ajustados de la mejor manera que puedas. Para terminar realizar un pequeño análisis de errores.

```python
# 3.1 Preparación de los datos para regresión
# En este apartado vamos a predecir la variable "alcohol" mediante modelos de regresión.

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Definimos X e y para regresión
X_reg = df.drop(columns=["alcohol"])
y_reg = df["alcohol"]

# Dividimos en train y test
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(
    X_reg, y_reg, test_size=0.2, random_state=42
)

# Escalado (solo para modelos que lo necesiten)
scaler_reg = StandardScaler()
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_reg.transform(X_test_reg)
```

```python
# 3.2 Entrenamiento de modelos de regresión

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

resultados_reg = {}

# Modelo 1: Linear Regression (requiere escalado)
lin_reg = LinearRegression()
scores_lin = cross_val_score(
    lin_reg,
    X_train_reg_scaled,
    y_train_reg,
    cv=5,
    scoring="neg_mean_squared_error"
)
resultados_reg["Linear Regression"] = -scores_lin.mean()

# Modelo 2: Random Forest Regressor (no requiere escalado)
rf_reg = RandomForestRegressor(random_state=42)
scores_rf = cross_val_score(
    rf_reg,
    X_train_reg,
    y_train_reg,
    cv=5,
    scoring="neg_mean_squared_error"
)
resultados_reg["Random Forest"] = -scores_rf.mean()

resultados_reg
```

```python
# 3.3 Comparación de modelos de regresión

import pandas as pd

df_reg_comparacion = pd.DataFrame.from_dict(
    resultados_reg, orient="index", columns=["MSE"]
).sort_values(by="MSE")

df_reg_comparacion
```

```python
# 3.4 Optimización del mejor modelo de regresión

from sklearn.model_selection import GridSearchCV

param_grid_reg = {
    "n_estimators": [100, 200],
    "max_depth": [None, 5, 10],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}

grid_rf_reg = GridSearchCV(
    estimator=rf_reg,
    param_grid=param_grid_reg,
    cv=5,
    scoring="neg_mean_squared_error",
    n_jobs=-1
)

grid_rf_reg.fit(X_train_reg, y_train_reg)

grid_rf_reg.best_params_, -grid_rf_reg.best_score_
```

```python
# 3.5 Evaluación final del modelo optimizado

from sklearn.metrics import mean_squared_error, r2_score

# Entrenamos el modelo final
mejor_rf_reg = RandomForestRegressor(
    n_estimators=grid_rf_reg.best_params_["n_estimators"],
    max_depth=grid_rf_reg.best_params_["max_depth"],
    min_samples_split=grid_rf_reg.best_params_["min_samples_split"],
    min_samples_leaf=grid_rf_reg.best_params_["min_samples_leaf"],
    random_state=42
)

mejor_rf_reg.fit(X_train_reg, y_train_reg)

# Predicciones
y_pred_reg = mejor_rf_reg.predict(X_test_reg)

# Métricas finales
mse_final = mean_squared_error(y_test_reg, y_pred_reg)
r2_final = r2_score(y_test_reg, y_pred_reg)

mse_final, r2_final
```

En este apartado se han entrenado y evaluado distintos modelos de regresión para predecir la variable “alcohol”. Tras preparar los datos y aplicar escalado cuando era necesario, se entrenaron dos modelos: Linear Regression y Random Forest Regressor. La comparación mediante validación cruzada mostró que Random Forest obtenía un error cuadrático medio (MSE) significativamente menor, por lo que se seleccionó como mejor candidato.

Posteriormente, se optimizó el modelo mediante GridSearchCV, obteniendo unos hiperparámetros muy similares a los valores por defecto, lo que indica que el modelo ya ofrecía un rendimiento sólido desde el inicio.
